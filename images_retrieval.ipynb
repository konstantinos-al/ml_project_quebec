```python
# libraries to use
from PIL import Image
import requests
from matplotlib import pyplot as plt
from io import BytesIO
```

```python
pre_url = "https://raw.githubusercontent.com/konstantinos-al/ml_project_quebec/main/Kaggle_helmets/images/"

end_url = 'https://raw.githubusercontent.com/konstantinos-al/ml_project_quebec/main/Kaggle_helmets/images/BikesHelmets0.png'

response = requests.get(end_url)
img = Image.open(BytesIO(response.content))

plt.imshow(img)
```

```python
prefix_url = "https://raw.githubusercontent.com/konstantinos-al/ml_project_quebec/main/Kaggle_helmets/images/"
number_of_imgs = 765

list_imgs = []

for i in range(0, number_of_imgs):
    i = str(i)
    var = "BikesHelmets" + i + ".png"
    var_url = prefix_url + var
    list_imgs.append(var_url)

for i in range(0, len(list_imgs)):
    response = requests.get(list_imgs[i])
    img = Image.open(BytesIO(response.content))
    plt.imshow(img)
```
